### **Há outliers nas distribuições das respectivas variáveis?**

É perceptível que algumas variáveis tem a sua dispersão afetada, examine as duas variáveis que iremos trabalhar, note que os dados estão concentrados em um dos lados, é possível presumir que este efeito é causado em razão da presença de outliers.

Para confirmar essa suposição podemos atuar de duas maneiras:

* **realizar uma análise descritiva das variáveis, por meio do comando .describe();**
* **plotar um boxplot para a variável em questão e identificar a presença de outliers;**

```
# Vamos abordar as duas maneiras, mas vamos começar pela análise descritiva
df[['price', 'minimum_nights']].describe() # Verificando como ficaram as métricas que importam na nossa análise
```

Após realizarmos a análise descritiva das variáveis de interesse confirmamos a presença de outliers em algumas delas, fazendo com que a distribuição destas sejam distorcidas, num momento oportuno podemos dar um melhor tratamento a elas.

```
# Vamos gerar uma cópia do nosso dataset para uma maior segurança nas manipulações que vamos realizar.
df_clean = df.copy() # Esse processo é realizado para que o banco de dados não seja corrompido
```

```
# Vamos analisar através do boxplot a variável price
# Vamos iniciar a figura do boxplot
fig, ax = plt.subplots(figsize=(15,3))

# Plotando o gráfico 
sns.set_theme(style="whitegrid")
sns.boxplot(x=df['price'])

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Anexando o título
plt.text(0, -0.6, 'Preços',
         fontsize=16,
         color='#787878',
         weight='bold')

# Mostrando o gráfico
plt.show()
```

O boxplot traz uma visão mais realista, mostrando a presença de inúmeros outliers, uma forma de lidar com eles é estabelecer um limite superior e inferior tanto para a variável price, quanto para a minimum_nights (estes limites funcionaram como um filtro), e verificar quantos valores acima destes existem, para assim conseguirmos entender a quantidade de outliers presentes nessa variável.



```
# Calcular as medidas necessárias para identificar outliers
Q1 = df['price'].quantile(q=0.25) # Esta linha corresponde ao primeiro quartil
Q3 = df['price'].quantile(q=0.75) # Esta linha corresponde ao terceiro quartil
IQR = Q3 - Q1 # Essa operação nos determina o interquartil, que nos auxiliará na fixação dos limites
limite_sup = Q3 + 1.5 * IQR # Limite superior
limite_inf = Q1 -1.5 * IQR # Limite inferior

# Mostrando os limites
print(f"O limite superior é: {limite_sup}.\nO limite inferior é: {limite_inf}.")

# Eliminar os outliers da amostra
df_clean = df[(df['price'] < limite_sup) & (df['price'] > limite_inf)] # Aqui estamos implementando os nossos limites, por meio de blocos condicionais
```

```
# Verificando se a limpeza acima foi bem executada
df_clean['price'].describe()
```

```
# Retirando valores iguais a zero na variável price, a qual pode trazer alguma distorção gráfica depois
df_clean.drop(df_clean[df_clean['price'] == 0].index, axis = 0, inplace=True)
```

```
# Verificando se o comando acima funcionou
df_clean.describe()
```

A variável price foi a primeira a ter um tratamento no que diz respeito a outliers, vejamos como ficou o boxplot após essa manutenção.

```
# Inicializando a figura do boxplot
fig, ax = plt.subplots(figsize=(15,3))

# Plotar o gráfico
sns.set_theme(style='whitegrid')
sns.boxplot(x= df_clean['price'])

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Anexando o título
plt.text(0, -0.6, 'Preços',
         fontsize=16,
         color='#787878',
         weight='bold')

# Mostrando o gráfico
plt.show()
```

Note como o aspecto do gráfico é outro, conseguimos determinar de maneira clara, os percentis, a mediana, os limites inferior e superior e ainda assim os outliers, vale salientar que a limpeza que fizessemos não foi para retirar todos os outliers, mas sim aqueles mais discrepantes em relação a nossa média, seja pra mais ou menos.

**Agora trataremos dos valores da variável minimum_nights.**

Realizaremos as mesmas operações praticamente, mudando um ou outro comando quando necessário para alcançar o resultado almejado.


```
# Inicializando a figura do boxplot
fig, ax = plt.subplots(figsize=(8,4))

# Plotar o gráfico
sns.set_theme(style='whitegrid')
sns.boxplot(x= df['minimum_nights'])

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Anexando o título
plt.text(0, -0.6, 'Noites mínimas',
         fontsize=16,
         color='#787878',
         weight='bold')

# Mostrando o gráfico
plt.show()
```

O gráfico acima está iteiramente distorcido em razão da presença de inúmeros outliers, veja que tem valores de noite mínima na casa de 1000, enquanto a sua média é de aproximadamente 5.
Portanto para esta variável vamos implementar a mesma técnica de limpeza realizada na variável price.

```
# Calculando as medidas necessárias para ontermos os outliers
Q_1 = df['minimum_nights'].quantile(q=0.25) # Definindo o primeiro quartil
Q_3 = df['minimum_nights'].quantile(q=0.75) # Definindo o terceiro quartil
IQR = Q_3 - Q_1 # Encontrando o interquartil
limite_sup_min = Q_3 + 1.5 * IQR # Definindo limte superior
limite_inf_min = Q_1 - 1.5 * IQR # Definindo limite inferior

# Printando os limites estabelecidos
print(f"O limite superior é: {limite_sup_min}.\nO limite inferior é: {limite_inf_min}.")
```

```
# Limpando o dataset
df_clean = df[(df['minimum_nights'] < limite_sup_min) & (df['minimum_nights'] > limite_inf_min)] # Implementando os limites de minimum_nights como um bloco condicional

# Vamos ver uma nova descrição
df_clean.minimum_nights.describe() # verificando se a limpeza surtil efeito
```

```
# Inicializando a figura do boxplot
fig, ax = plt.subplots(figsize=(8,4))

# Plotar o gráfico
sns.set_theme(style='whitegrid')
sns.boxplot(x= df_clean['minimum_nights'])

# Remover grids e eixos
ax.spines['right'].set_visible(False)
ax.spines['left'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.spines['bottom'].set_visible(False)

# Anexando o título
plt.text(0, -0.6, 'Noites mínimas',
         fontsize=16,
         color='#787878',
         weight='bold')

# Mostrando o gráfico
plt.show()
```

No gráfico acima já conseguimos obter uma visualização melhor após a limpeza efetuada, ou seja, qual é a realidade dos nossos dados, apenas estipulando novos limites, mas lembre-se isso não significa que removemos os nossos outliers.

### **Qual a correlação existente entre as variáveis?**

Vamos determinar a correlação entre duas variáveis, neste ponto é necessário ficar atento para não confundir correlação com causalidade, assim a título de experimentação vamos correlacionar duas variáveis que irão interagir de alguma maneira, onde poderam ter uma correlação positiva ou negativa mas isto não significa que o aumento de uma está causando o aumento na outra e vice-versa.

Essa pode ser determinada por meio do coeficiente de correlação, muito utilizado em machine learning na determinação de features, que determinará a sua intensidade, para estabelecermos esse coeficiente vamos criar uma matriz de correlação e gerar um heatmap para visualizarmos melhor o resultado.

```
# criar uma matriz de correlação
corr = df_clean[['id',	'name',	'neighbourhood',	'latitude',	'longitude',	'room_type',	'price',	'minimum_nights']].corr()

# mostrar a matriz de correlação
display(corr) # Usamos o próprio pandas para fazer essa correlação
```

```
# plotar um heatmap a partir das correlações
sns.heatmap(corr, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True, vmin=0.0, vmax=1.0, center=0);
```

Podemos perceber acima que existe pouca correlação ou quase nenhuma entre as variáveis, e que as variáveis room_type e neighbourhood foram deixadas de lado por serem colunas categóricas, por isso necessitamos convertê-las em variáveis numéricas para que possamos estabelecer uma ligação entre eles e as demais. Para isso precisamos converter estas variáveis em binários, ou seja, valores numéricos.

```
#Importando as bibliotecas
from sklearn.preprocessing import OneHotEncoder # Por meio destas bibliotecas conseguimos realizar essa tarefa
from sklearn.preprocessing import LabelEncoder
```

Para essa tarefa vamos usar o labelencoder atributo da biblioteca sklearn a qual tem alguns parâmetros interessnates para essa tarefa, onde num primeiro momento vamos utilizar o próprio labelenconder que irá converter os categóricos em numéricos, entretanto sem atribuir o mesmo peso de importância as novas entradas, deste modo vamos aplicar o onehotencoder o qual irá padronizar estas de maneira que todas as novas atribuições tenham o mesmo peso. Algo importante também é criar uma base de dados específica para essa tarefa visto que estamos modificando o nosso dataset.

```
# Vamos realizar nova cópia do dataset
df_corr = df_clean.copy()

# A variável neighbourhood será convertida em inteiro
label_encod = LabelEncoder()
int_encod = label_encod.fit_transform(df_corr['neighbourhood'])

# Agora pra binário
onehot_encoder = OneHotEncoder(sparse=False)
int_encod = int_encod.reshape(len(int_encod), 1)
df_corr = df_corr
df_corr['neighbourhood'] = onehot_encoder.fit_transform(int_encod)

# Transformando room_type
label_encod = LabelEncoder()
int_encod = label_encod.fit_transform(df_corr['room_type'])

# Agora pra binário
onehot_encoder = OneHotEncoder(sparse=False)
int_encod = int_encod.reshape(len(int_encod), 1)
df_corr['room_type'] = onehot_encoder.fit_transform(int_encod)
```

```
# Visualizando o nosso df
df_corr.sample(20) # Checando se a transformação logrou êxito
```

```
# Vamos montar uma nova matriz de correlação e comparar o resultado
correlacao = df_corr.corr() # Atribuindo um valor para a nossa variável correlação

# Analisando visualmente a matriz de correlção
display(correlacao)
```

```
# plotar um heatmap a partir das correlações
sns.heatmap(correlacao, cmap='RdBu', fmt='.2f', square=True, linecolor='white', annot=True, vmin=0.0, vmax=1.0, center=0);
```

Com a transformação de algumas variáveis categóricas em numéricas conseguimos realizar uma análise mais profunda e percebemos que há uma correlação mais forte entre price, minimum_nights e room_type, entre as quais pode haver uma causalidade, mas só por meio desta análise não conseguimos cravar isso.

### **Qual o tipo de imóvel mais alugado no Airbnb?**

A variável room_type nos dá informação em relação ao tipo de imóvel que está sendo alugado, seja um quarto, um apartamento inteiro, uma casa, e desta forma queremos saber qual destes é o mais alugado. Para realizar essa tarefa vamos utilizar o comando value_counts().
